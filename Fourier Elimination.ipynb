```python
x=(((2,3,-1,1,1,4,-4),(1,-1,1,-1,2,-2,2),(-1,1,-1,1,-2,2,-2),(-2,-1,2,-2,-1,-3,3)),(5,4,-4,-3))
y=len(x[1])
```

```python
import numpy as np
A=np.asarray(x[0],dtype=np.float)
b=np.asarray(x[1],dtype=np.float)
b=b.reshape(y,1)
p=A.shape
m=p[0]
n=p[1]
V=np.append(A,b,1)
```

# fourier function

```python
finalA=[V]
def a():
    global A
    global b
    global m,n,opt,nonzeros,x
    def samesign():
        global m,n,opt,finalA,nonzeros,x
        for col in range(n):
            count=[]
            counter=[]
            for row in range(m):
                if A[row,col]>0:
                    count.append(1)
                elif A[row,col]<0:
                    counter.append(1) 
            if len(count)==m:
                for i in range(m):
                    print(b[i],A[i,col])
                    b[i]=b[i]/A[i,col]
                    print(b[i])
                y=min(b)
                print(y)
                k=len(finalA)
                print(k,finalA)
                if k!=1:       #if more than one step answer
                    opt=[]
                    z=n+1
                    for i in range(z):#z=n+1
                        opt.append(0)
                    opt[col+1]=y #+1 because we will be using this opt for a step behind the usual back substitution.
                    u=0
                    while u<=col:
                        opt[u]=1
                        u=u+1
                    opt=np.asarray([opt],dtype=np.float)
                    fin=-2
                    while fin>=(-1)*k:
                        x=np.asarray(finalA[fin])
                        print(x)
                        bx=x[:,-1]
                        x=np.delete(x,-1,1)
                        print(x,bx)
                        p=x.shape
                        bx=bx.reshape(p[0],1)
                        fin=fin-1
                        for finsub in range(p[0]):
                            print(p,x[finsub,:],opt,opt.shape)
                            x[finsub,:]=(x[finsub,:]*opt)
                        ub=[]
                        lb=[]
                        for r in range(p[0]):
                            for c in range(1,z):
                                bx[r]=bx[r]-x[r,c]
                            if x[r,0]!=0:    
                                bx[r]=bx[r]/x[r,0]
                                if x[r,0]>0:
                                    ub.append(bx[r])
                                if x[r,0]<0:
                                    lb.append(bx[r])
                        l=max(lb,default=0) 
                        u=min(ub,default=0)
                        if l>u:
                            x="infeasible"
                            return x
                        y=(l+u)/2
                        opt[0,0]=y
                        print(opt.shape,V.shape)
                        if opt.shape[1]!=V.shape[1]-1:
                            z=z+1
                            one=[[1]]
                            one=np.asarray(one)
                            opt=np.append(one,opt,axis=1)                                   
                else:                    # if first step answer
                    for i in range(n):
                        opt.append(0)
                    opt[col]=y
                opt=np.asarray(opt,dtype=np.float)       
                print(opt)
                print("optimal solution is ",opt)
                return opt
                break
            if len(counter)==m:  #COUNTER
                for i in range(m):
                    b[i]=b[i]/A[i,col]
                y=max(b)
                print(y,col)
                opt=[]
                if k!=1:       #if more than one step answer
                    opt=[]
                    z=n+1
                    for i in range(z):#z=n+1
                        opt.append(0)
                    opt[col+1]=y #+1 because we will be using this opt for a step behind the usual back substitution.
                    u=0
                    while u<=col:
                        opt[u]=1
                        u=u+1
                    opt=np.asarray([opt],dtype=np.float)
                    fin=-2
                    while fin>=(-1)*k:
                        x=np.asarray(finalA[fin])
                        bx=x[:,-1]
                        x=np.delete(x,-1,1)
                        p=x.shape
                        bx=bx.reshape(p[0],1)
                        fin=fin-1
                        for finsub in range(p[0]):
                            x[finsub,:]=(x[finsub,:]*opt)
                        ub=[]
                        lb=[]
                        for r in range(p[0]):
                            for c in range(1,z):
                                bx[r]=bx[r]-x[r,c]
                            if x[r,0]!=0:    
                                bx[r]=bx[r]/x[r,0]
                                if x[r,0]>0:
                                    ub.append(bx[r])
                                if x[r,0]<0:
                                    lb.append(bx[r])
                        l=max(lb,default=0) 
                        u=min(ub,default=0)            
                        if l>u:
                            x="infeasible"
                            return x
                        y=(l+u)/2
                        opt[0,0]=y
                        print(opt.shape,V.shape)
                        if opt.shape[1]!=V.shape[1]-1:
                            z=z+1
                            one=[[1]]
                            one=np.asarray(one)
                            opt=np.append(one,opt,axis=1)
                                                    
                else:                    # if first step answer
                    for i in range(n):
                        opt.append(0)
                    opt[col]=y
                opt=np.asarray(opt,dtype=np.float)       
                print(opt)
                print("optimal solution is ",opt)
                return opt
                break
            print(len(counter),m)    
                
    opt=[]
    x="nil"
    samesign()
    if x=="infeasible":
        return x
    if n==1:
        ub=[]
        lb=[]
        for r in range(m):
            if A[r,0]!=0:
                b[r]=b[r]/A[r,0]
                if A[r,0]>0:
                    ub.append(b[r])
                if A[r,0]<0:
                    lb.append(b[r])
        l=max(lb,default=0) 
        u=min(ub,default=0)
        if l>u:
            x="infeasible"
            return x
        y=(l+u)/2
        opt=[]
        z=n+1
        col=0
        for i in range(z):#z=n+1
            opt.append(0)
        opt[col+1]=y #+1 because we will be using this opt for a step behind the usual back substitution.
        u=0
        while u<=col:
            opt[u]=1
            u=u+1
        opt=np.asarray([opt],dtype=np.float)
        k=len(finalA)
        fin=-2
        while fin>=(-1)*k:
            x=np.asarray(finalA[fin])
            print(x)
            bx=x[:,-1]
            x=np.delete(x,-1,1)
            print(x,bx)
            p=x.shape
            bx=bx.reshape(p[0],1)
            fin=fin-1
            for finsub in range(p[0]):
                print(p,x[finsub,:],opt,opt.shape)
                x[finsub,:]=(x[finsub,:]*opt)
            nonzeros=[]    
            for r in range(p[0]):
                for c in range(1,z):
                    bx[r]=bx[r]-x[r,c]
                if x[r,0]!=0:    
                    bx[r]=bx[r]/x[r,0]
                    nonzeros.append(b[x])
            y=min(nonzeros)
            opt[0,0]=y
            print(opt.shape,V.shape)
            if opt.shape[1]!=V.shape[1]-1:
                z=z+1
                one=[[1]]
                one=np.asarray(one)
                opt=np.append(one,opt,axis=1)
                                                        
    j=0 
    while j<=n:
        if len(list(opt))>0:
            print(opt)
            return opt
        Ap=[]
        An=[]
        bp=[]
        bn=[]
        for i in range(m):
            count=[]
            for j in range(n):
                if b[i]!=0:
                    break
                else:
                    count.append(0)
            if len(count)==n:
                np.delete(A,i,0)
                m=m-1
        for i in range(m):
            if A[i,j]==0:
                j=j+1
                break
        for i in range(m):        
            if A[i,j]>0:
                Ap.append(A[i,:]/A[i,j])
                bp.append(b[i]/A[i,j])
            else:
                An.append(A[i,:]/abs(A[i,j]))
                bn.append(b[i]/abs(A[i,j]))
        Ap=np.asarray(Ap)
        Ap=np.delete(Ap,0,1)#nth column
        print(Ap)
        An=np.asarray(An)
        An=np.delete(An,0,1)#n column
        bp=np.asarray(bp)
        bn=np.asarray(bn)
        Ap=Ap*(-1) 
        bn=bn*(-1)
        Ap=np.append(Ap,bp,1)
        An=np.append(An,bn,1)
        p=Ap.shape
        q=An.shape
        finalAsub=[]             #final A sub matrix(including b)
        for xp in range(p[0]):
            for yn in range(q[0]):
                z=An[yn]-Ap[xp]
                finalAsub.append(z)
            count=[]
            print(z)
            for i in range(q[1]):
                if z[i]==0:
                    count.append(0)
            print(len(count),q[1])        
            if len(count)==q[1]:
                del(finalAsub[-1])       
        finalA.append(finalAsub)        
        A=np.asarray(finalAsub)        #segregating finalAsub for next iteration
        p=A.shape
        print(p)
        m=p[0]
        n=p[1]
        b=(-1)*A[:,-1]
        b=b.reshape(p[0],1)
        A=np.delete(A,-1,1)
        p=A.shape
        print(p)
        m=p[0]
        n=p[1]
        print(A,b,m,n)
        a()
        
       
            
        
        
        
        
```
